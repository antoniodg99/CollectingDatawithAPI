```python
# URL dell' API
api_url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

#Scarico i dati
response = requests.get(api_url)

#Verifichiamo se la conession è data a buon fine
if response.ok:
    data = response.json()
    
# In questo caso non stampo i dati in quanto sono tanti
# utilizzo un df e uso df.head

#Inserisco i dati nel df
df = pd.DataFrame(data)

#Stampo le prime righe del df
df.head()
```

```python
def get_number_of_jobs_T(technology):
    tech_lower = technology.lower()
    number_of_jobs = 0      # Creato un contatore a 0
    for job in data:
        #Estrai dal dizionario JOB i valori della chiave "Key Skills"
        #Altrimenti lascia la stringa vuota (Chiave, "")
        skills = job.get("Key Skills", "")
        if tech_lower in skills.lower():   #Se trovi tech_lower nella chiave "Key Skills"
            number_of_jobs += 1            # Aggiungi +1 al contatore number of jobs
    return technology, number_of_jobs
```

Calling the function for Python and checking if it works.


```python
get_number_of_jobs_T("Power BI")
```

#### Write a function to find number of jobs in US for a location of your choice


```python
#Facciamo la stessa operazione per le località
def get_number_of_jobs_L(location):
    loc_lower = location.lower()
    number_of_location = 0
    for job in data:
        job_locations = job.get("Location","")
        if loc_lower in job_locations.lower():
            number_of_location += 1
    return location, number_of_location
```

Call the function for Los Angeles and check if it is working.


```python
get_number_of_jobs_L("Los Angeles")
```

Create a python list of all technologies for which you need to find the number of jobs postings.


```python
# Creo lista vuota
technologies_list = set()

# Itero tutti i dizionari nella lista data
for job in data:
    
    #Ciclo tutte le key skills eliminando suddividendo per "|"
    for skills in job.get("Key Skills","").split("|"):  
        #Rimuovo spazi vuoti e rendo tutto minuscolo
        skills = skills.strip().lower()
    if skills:
        #Aggiungo la skill
        technologies_list.add(skills)


#Check
print("Ci sono",len(technologies_list),"tecnologie diverse")

#Nuova funzione
def get_all_jobs_count_by_T(technologies):
    results = []
    for tech in technologies:
        tech_name, count = get_number_of_jobs_T(tech)
        results.append({"Technology": tech_name,
                       "Count": count}
                     )
    return results

job_counts = get_all_jobs_count_by_T(technologies_list)

print(job_counts)
```
